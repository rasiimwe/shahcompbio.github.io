```python

model = fit_results['model'].model

state_sequence = np.zeros((model.num_segments,), dtype=np.int64)
allele_state_prob = np.zeros((model.num_alleles, model.num_alleles))
allele_states = np.zeros((2,), dtype=np.int64)
new_p_breakpoint = np.zeros((model.num_breakpoints, model.num_brk_states, model.num_alleles, model.num_alleles))

def _argmax2(values, indices):
    vmax = values.min() - 1.
    indices[0] = 0
    indices[1] = 0
    for i in range(values.shape[0]):
        for j in range(values.shape[1]):
            if values[i, j] > vmax:
                vmax = values[i, j]
                indices[0] = i
                indices[1] = j

for k in range(model.num_breakpoints):
    allele_state_prob[:] = 0.
    for s_b in range(model.num_brk_states):
        for v_1 in range(2):
            for v_2 in range(2):
                allele_state_prob[v_1, v_2] += model.p_breakpoint[k, s_b, v_1, v_2]

    _argmax2(allele_state_prob, allele_states)

    for s_b in range(model.num_brk_states):
        for v_1 in range(2):
            for v_2 in range(2):
                new_p_breakpoint[k, s_b, allele_states[0], allele_states[1]] += model.p_breakpoint[k, s_b, v_1, v_2]

model.p_breakpoint = new_p_breakpoint

```
